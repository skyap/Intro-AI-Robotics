```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)
x = np.linspace(0,100,101)
y = 3.4*x+9+np.random.random(101)*200
```

```python
fig = plt.subplots(figsize=(20,10))
plt.scatter(x,y)
plt.show()
```

```python
n = len(x)
mx = 1/n*sum(x)
my = 1/n*sum(y)
m = sum((x-mx)*(y-my))/sum((x-mx)**2)
c = my - m*mx
print(m,c)
```

```python
py = m*x+c
```

```python
mse = sum((py-y)**2) / n
mse
```

```python


fig = plt.subplots(figsize=(20,10))
plt.scatter(x,y)
plt.plot(x,py)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.optimizers import *
```

```python
xx = Input(shape=(1,))
yy = Dense(units = 1)(xx)
model = Model(inputs=xx,outputs=yy)
model.compile(optimizer=Adam(0.1),loss="mse")
```

```python

model.fit(x.reshape(-1,1),y.reshape(-1,1),epochs=1000, batch_size=10,verbose=1)
weights,biases = model.layers[1].get_weights()
```

```python
pyy = weights[0]*x+biases
```

```python
sum((pyy-y)**2)/n
```

```python
print(weights,biases)
```

```python


fig = plt.subplots(figsize=(20,10))
plt.scatter(x,y)
plt.plot(x,pyy,color="green",label="neural network")
plt.plot(x,py,color="red",label="analytic")
plt.legend()
plt.show()
```
